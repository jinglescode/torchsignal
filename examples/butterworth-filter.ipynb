```python
from torchsignal.datasets.generate import generate_signal

signal_original = generate_signal(length_seconds=2, sampling_rate=1000, frequencies_list=[8,15])
```

```python
from torchsignal.preprocessing.bandpass.butterworth import butter_bandpass_filter_signal_1d

signal_bandpassed = butter_bandpass_filter_signal_1d(signal_original, lowcut=5, highcut=10, sample_rate=1000, order=4, verbose=True)
```

```python
from numpy import dot
from numpy.linalg import norm

signal_8hz = generate_signal(length_seconds=2, sampling_rate=1000, frequencies_list=[8])
cosine_similarity = dot(signal_8hz, signal_bandpassed)/(norm(signal_8hz)*norm(signal_bandpassed))
print(cosine_similarity)
```
